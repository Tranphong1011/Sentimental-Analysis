### Scrapping thông tin Restaurant


```python
import os

directory = 'Scrapping data/'

Restaurant = []

for filename in os.listdir(directory):
    if filename.endswith('.txt') and filename != 'Restaurant Info.txt':  
        Restaurant.append(filename) 

# Print the list of file names
print(Restaurant)

```

```python
with open('Scrapping data/Restaurant Info.txt', 'r', encoding='utf-8') as file:
    restaurant_info = file.readlines()  # Read all lines into a list
restaurant_info
```

```python
restaurant_names = []
addresses = []
operating_times = []
prices = []

for i in range(0, len(restaurant_info), 4):
    restaurant_names.append(restaurant_info[i].strip())

for i in range(1, len(restaurant_info), 4):
    addresses.append(restaurant_info[i].strip().replace('Address: ', ''))

for i in range(2, len(restaurant_info), 4):
    operating_times.append(restaurant_info[i].strip().replace('Time: ', ''))

for i in range(3, len(restaurant_info), 4):
    price = restaurant_info[i].strip().replace('Price: ', '').replace(' VND', '')  # Corrected line
    prices.append(price)
print("Restaurant Names:", restaurant_names)
print("Addresses:", addresses)
print("Operating Times:", operating_times)
print("Prices:", prices)
```

```python
import pandas as pd

data = {
    'ID': range(1623, 1623 + len(restaurant_names)),
    'Restaurant': restaurant_names,
    'Address': addresses,
    'Time': operating_times,
    'Price': prices
}

df = pd.DataFrame(data)

df
```

```python
import re

name = []
rating = []
reviews = []
id_restaurant = []
time = []
for filename in Restaurant:
    with open(os.path.join(directory, filename), 'r', encoding='utf-8') as file:
        lines = file.readlines()
    data_without_newline = [item.replace('\n', '') for item in lines]
    # Define a regular expression pattern to match date and time strings
    date_time_pattern = r'\d{2}-\d{2}-\d{4} \d{2}:\d{2}'

    # Extract all date and time strings from the list
    date_times = re.findall(date_time_pattern, ' '.join(data_without_newline))
    time.extend(date_times)
    result = []
    sublist = []

    for item in data_without_newline:
        sublist.append(item)
        if item in date_times:
            result.append(sublist)
            sublist = []
            list_2 = date_times[date_times.index(item):]
    if sublist:  
        result.append(sublist)
    for sublist in result:
        name.append(sublist[0])
        rating.append(sublist[1])
        reviews.append(' '.join(sublist[2:-1]))
        id_restaurant.append(df.loc[df['Restaurant'] == filename.replace(".txt",""), 'ID'].values[0])
```
